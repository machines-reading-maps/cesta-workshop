# Labels over Time

This example notebook show how you can create simple plots based on the metadata files in the folder `01_data_preparation`. We will use the the file `map-metadata.json` to generate a scatterplot of the number of labels over time.

```python
import json

discarded = 0

x = []

y = []

with open('../01_data_preparation/output/map-metadata.json') as f:
  data = json.load(f)

  for obj in data:
    if 'date_normalized' in obj and 'num_labels' in obj:
      x.append(obj['date_normalized'])
      y.append(obj['num_labels'])

    else:
      discarded += 1
    
  print('Total discarded: ' + str(discarded) + ' of ' + str(len(data)))
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

plt.subplots(figsize=(10, 8))

sns.regplot(
  x=x, 
  y=y, 
  ci=None, 
  scatter_kws={ 's':1 },
  line_kws={'color': 'red', 'label': 'Trend'})

plt.yscale('log')
plt.legend()
plt.show()
```
